```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import logging
import pymongo
import os
from pymongo import MongoClient
from dotenv import load_dotenv
```

```python
save_dir = "images/"
if not os.path.exists(save_dir):  # makes directory if not already exists
    os.makedirs(save_dir)
```

```python
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/123.0.0.0 Safari/537.36"
    ),
    "Accept-Language": "en-US,en;q=0.9",
}
```

```python
query = "Virat Kohli"
response = requests.get(f"https://in.images.search.yahoo.com/search/images;_ylt=Awr1TYm6B8toEgIAKcu7HAx.;_ylu=Y29sbwNzZzMEcG9zAzEEdnRpZAMEc2VjA3BpdnM-?p={query}&fr2=piv-web&type=E210IN826G0&fr=mcafee") # f helps in setting the query format to string
# response will get or scrape the whole web page and give you the image.
```

```python
soup = bs(response.content,'html.parser') #beautifying and storing it inside a variable
```

```python
images_tag = soup.find_all("img")
```

```python
len(images_tag)
```

```python
del images_tag[0] #deleting the first one cause the first one is not an image but rather a header
```

```python
img_data_mongo = []
for i in images_tag:
    # Check if the 'src' attribute exists
    if 'src' in i.attrs:
        image_url = i['src']
        try:
            image_data = requests.get(image_url).content
            mydict = {"index": image_url, "image": image_data}
            img_data_mongo.append(mydict)
            
            # Save image to file
            with open(os.path.join(save_dir, f"{query}_{images_tag.index(i)}.jpg"), "wb") as f:
                f.write(image_data)
        except Exception as e:
            print(f"Error downloading {image_url}: {e}")
    else:
        print(f"Skipping image, 'src' attribute not found: {i}")
        
```

```python
load_dotenv(dotenv_path=r"C:\Users\ifham\Downloads\project\ .env")
```

```python
mongo_uri = os.getenv("MONGO_URI")
client = MongoClient(mongo_uri)
db = client['image_scrape'] # this is a database defined
coll_image = db['image_scrape'] # this is a collection defined
coll_image.insert_many(img_data_mongo)
```
